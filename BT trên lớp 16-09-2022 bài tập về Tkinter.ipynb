```python
from tkinter import * 
from tkinter import ttk
from tkinter import messagebox

def selection(event): 
    messagebox.showinfo('Dũ liệu được chọn: ', combobox.get())
#     messagebox.showinfo('Dũ liệu được chọn: ', variable.get())

root = Tk()
root.title('Combobox Demo')
root.geometry('300x300')

# values = ['Python', 'Java', 'Android', 'Angula']
values = [f'Tháng {i}' for i in range(1,13,1)]
variable = StringVar()

# combobox = ttk.Combobox(root, values = values, textvariable = variable, state = 'readonly')
# combobox.pack()
# combobox.set(values[1])
# combobox.bind('<<ComboboxSelected>>', selection)

frame = Frame(root, height = 100, width = 200, borderwidth = 1)
frame.pack(side = BOTTOM)
combobox = ttk.Combobox(frame, values = values, textvariable = variable, state = 'readonly')
combobox.pack()
combobox.set(values[1])
combobox.bind('<<ComboboxSelected>>', selection)


button = Button(frame, text = 'Summit')
button.bind('<Button>', selection)
button.pack()

root.mainloop()
```
